```python
import pandas as pd

targetUrl_iris="https://raw.githubusercontent.com/gundaminpde/2022/main/iris.csv"


# 아이리스 데이터를 불러옵니다.
df = pd.read_csv(targetUrl_iris)
```

```python

df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 그래프로 확인해 봅시다.
sns.pairplot(df.iloc[:,1:6], hue='Species');
plt.show()
```

```python
# 속성을 X, 클래스를 y로 저장.
X = df.iloc[:,1:5]
y = df.iloc[:,5]

# X와 y의 첫 5줄을 출력.
print(X[0:5])
print(y[0:5])
```

```python
# 원-핫 인코딩 처리를 합니다.
y = pd.get_dummies(y)

# 원-핫 인코딩 결과를 확인합니다.
print(y[0:5])
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# 모델 설정
model = Sequential()
model.add(Dense(12,  input_dim=4, activation='relu'))
model.add(Dense(10,  activation='relu'))
model.add(Dense(3, activation='softmax'))

# 요약
model.summary()

# 모델 컴파일
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# 모델 실행
history=model.fit(X, y, epochs=30, batch_size=5)

```

```python
## 그래서~~~ 다음과 같다.


from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

import pandas as pd
import matplotlib.pyplot as plt

# 아이리스 데이터를 불러옵니다.

targetUrl_iris="https://raw.githubusercontent.com/gundaminpde/2022/main/iris.csv"

df = pd.read_csv(targetUrl_iris)



print(df.iloc[:,1:6].head()) ### 1번째 column은 숫자 index라서 제외함.

# 속성을 X, 클래스를 y로 저장합니다.
X = df.iloc[:,1:5]
y = df.iloc[:,5]

# 원-핫 인코딩 처리를 합니다.
y = pd.get_dummies(y)

# 모델 설정
model = Sequential()
model.add(Dense(12,  input_dim=4, activation='relu'))
model.add(Dense(10,  activation='relu'))
model.add(Dense(3, activation='softmax'))
model.summary()

# 모델 컴파일
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# 모델 실행
history=model.fit(X, y, epochs=30, batch_size=5)


```
