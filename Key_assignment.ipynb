```python
from keras.models import Sequential
from keras.layers import Dense
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

targetUrl="https://raw.githubusercontent.com/gundaminpde/test1/main/test2.csv"

df = pd.read_csv(targetUrl)

y = df.iloc[:,4]
df =df.iloc[:,1:4]

##normalization_df=(df-df.mean(numeric_only=True))/df.std(numeric_only=True)

normalization_df=(df-df.mean())/df.std()

print(normalization_df.head())

x=normalization_df

print(x.head())
print(y.head())

y=pd.get_dummies(y)

print(y.head())
print(y.tail())


### del normalization_df
```

```python
model = Sequential()
model.add(Dense(36, input_dim=3, activation='sigmoid'))
model.add(Dense(33, activation='relu'))
model.add(Dense(6,activation='softmax'))
model.summary()
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])


model.fit(x,y, epochs=100, batch_size=20)
```
