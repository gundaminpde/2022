```python
### 홀수 짝수 구분을 학습1###

from keras.models import Sequential
from keras.layers import Dense
import numpy as np

x = np.array(range(1,1000))
y = np.array([[z%2,1-(z%2)] for z in range(1,1000)])

x_2=np.array([10,301, 30])

model = Sequential()

model.add(Dense(10, input_dim=1, activation='sigmoid'))
model.add(Dense(30, activation='sigmoid'))
####model.add(Dense(40, activation='sigmoid'))
model.add(Dense(2,activation='softmax'))

model.compile(loss='mse', optimizer='sgd')
          
model.fit(x,y,epochs=40, batch_size=30)

loss = model.evaluate(x,y) 

print('loss:', loss)
          
y_predict = model.predict(x_2)
print(y_predict)

```

```python
### 홀수 짝수 구분을 학습2###

from keras.models import Sequential
from keras.layers import Dense

import numpy as np

import matplotlib.pyplot as plt


x = np.array(range(1,1000))
y = np.array([[z%2,1-(z%2)] for z in range(1,1000)])

x_2=np.array([10,301, 30])

model = Sequential()

model.add(Dense(10, input_dim=1, activation='sigmoid'))
model.add(Dense(30, activation='sigmoid'))
###model.add(Dense(40, activation='sigmoid'))
model.add(Dense(2,activation='softmax'))

model.compile(loss='mse', optimizer='sgd')
          
history = model.fit(x,y,epochs=30, batch_size=30)

loss = model.evaluate(x,y) 

print('loss:', loss)
          
y_predict = model.predict(x_2)
print(y_predict)

y_loss = history.history['loss']

x_len = np.arange(len(y_loss))

plt.plot(x_len,y_loss)
```

```python
### 홀수 짝수 구분을 학습3###

from keras.models import Sequential
from keras.layers import Dense

import numpy as np

import matplotlib.pyplot as plt


x = np.array(range(1,1000))
y = np.array([z%2 for z in range(1,1000)])

x_2=np.array([10,301, 30])

model = Sequential()

model.add(Dense(10, input_dim=1, activation='relu'))
model.add(Dense(10, activation='sigmoid'))
model.add(Dense(10, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='mse', optimizer='sgd')
          
history = model.fit(x,y,epochs=40, batch_size=30)

loss = model.evaluate(x,y) 

print('loss:', loss)
          
y_predict = model.predict(x_2)
print(y_predict)

y_loss = history.history['loss']

x_len = np.arange(len(y_loss))

plt.plot(x_len,y_loss)
```

```python
### 20.22 보다 큰 수 구분하기###

from keras.models import Sequential
from keras.layers import Dense

import numpy as np

import matplotlib.pyplot as plt


x = np.array([ x/100  for x in range(1,5000)])
y = np.array([[int(z/100 > 20.22),1-int(z/100 > 20.22)] for z in range(1,5000)])

x_2=np.array([10,24.6, 30])

model = Sequential()

model.add(Dense(10, input_dim=1, activation='sigmoid'))
model.add(Dense(40, activation='sigmoid'))
model.add(Dense(2,activation='softmax'))

model.compile(loss='mse', optimizer='sgd')
          
history = model.fit(x,y,epochs=150, batch_size=30)

loss = model.evaluate(x,y) 

print('loss:', loss)
          
y_predict = model.predict(x_2)
print(y_predict)

y_loss = history.history['loss']

x_len = np.arange(len(y_loss))

plt.plot(x_len,y_loss)
```
