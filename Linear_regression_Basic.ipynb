```python
#import sklearn
import numpy as np
```

```python
# 항상 결과를 동일하게 유지하기 위해
np.random.seed(42)

# 그래프를 위해서
#%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rc('axes', labelsize=14)
mpl.rc('xtick', labelsize=12)
mpl.rc('ytick', labelsize=12)
```

```python
### data set 만들기

import numpy as np # 사실 이건 중복이다.

X = 2 * np.random.rand(100, 1)
y = 4 - 3* X + 4 * X**2 + np.random.randn(100, 1)
```

```python
## data_set 그리기

plt.plot(X, y, "b.")
plt.xlabel("$x_1$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.axis([0, 2, 0, 15])

plt.show()
```

```python
## 사실 이게 다야

from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(X, y)
lin_reg.intercept_, lin_reg.coef_
```

```python
X_new = np.array([[0], [2]])

lin_reg.predict(X_new)
```

```python
X_new=np.linspace(0, 2, 100).reshape(100, 1)
y_new = lin_reg.predict(X_new)


plt.plot(X, y, "b.")
plt.plot(X_new, y_new, "r-", linewidth=2, label="Predictions")
plt.xlabel("$x_1$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.legend(loc="upper left", fontsize=14)
plt.axis([0, 2, 0, 15])
plt.show()
```

```python
import numpy as np
import numpy.random as rnd

np.random.seed(42)
```

```python
m = 20
X = 8 * np.random.rand(m, 1) - 4
y =  X**2 + 2*X + 2 + 3* np.random.randn(m, 1)
```

```python
import matplotlib.pyplot as plt

plt.plot(X, y, "b.")
plt.xlabel("$x_1$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.axis([-4, 4, -10, 40])
plt.show()
```

```python
#7차식으로 만들기 위해 feature를 확장합니다.

from sklearn.preprocessing import PolynomialFeatures

poly_features = PolynomialFeatures(degree=30, include_bias=False)
X_poly = poly_features.fit_transform(X)
X[0]
```

```python
X_poly[0]
```

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(X_poly, y)
lin_reg.intercept_, lin_reg.coef_
```

```python


X_new=np.linspace(-4, 4, 100).reshape(100, 1)
X_new_poly = poly_features.transform(X_new)
y_new = lin_reg.predict(X_new_poly)
plt.plot(X, y, "b.")
plt.plot(X_new, y_new, "r-", linewidth=2, label="Predictions")
plt.xlabel("$x_1$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.legend(loc="upper left", fontsize=14)
plt.axis([-4, 4, -10, 40])
plt.show()
```
